# Описание задачи

- Оператор связи борется с оттоком клиентов, предлагая промокоды и специальные условия всем, кто планирует отказаться от его услуг. 
- Требуется построить модель, которая будет предсказывать, разорвёт ли абонент договор.
- Оператору требуется определять клиентов, планирующих уходить, но при этом минимизировать расходы, сократив предложение спецусловий тем клиентам, которые уходить не планируют.
- В качестве метрики качества модели оператор требует использовать `ROC-AUC`.
- Значение метрики на тестовой выборке должно быть более **0.85**


**Описание услуг оператора**

Оператор предоставляет два основных типа услуг: 
- Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL), или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).


Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

## Обучим и протестируем модели методом кросс-валидации

### Обучим и протестируем модель `LogisticRegression` методом `кросс-валидации`

#### Зададим неизменяемые гиперпараметры модели

```python
lr = LogisticRegression(
    solver = 'lbfgs', 
    random_state = RANDOM_STATE,
    n_jobs = -1, 
    max_iter = 1000
)
```

#### Зададим функцию, балансирующую классы

```python
cat_features = cat_features_determination(features)
cat_features
```

```python
balancer = SMOTENC(random_state=RANDOM_STATE,
                  categorical_features = cat_features)
```

#### Зададим трансформер столбцов  и упакуем в него кодировщики количественных и категориальных признаков

```python
lr_column_transformer = make_column_transformer(

    (RobustScaler(), make_column_selector(dtype_include='number')),
    
    (OneHotEncoder(drop='first', handle_unknown='ignore'), 
     make_column_selector(dtype_include='object')),
    
    remainder='passthrough')
```

#### Сконструируем конвейер для предобработки данных и обучения модели

```python
lr_pipe = Pipeline(steps=[
    ("balancer", balancer),
    ("lr_column_transformer", lr_column_transformer),
    ("lr", lr)])
print(lr_pipe)
```

#### Зададим диапазоны гиперпараметров для поиска оптимальной метрики `AUC-ROC` методом `кросс-валидации`

```python
lr_param_grid = {
"lr__C" : [0.4] #, 0.2, 0.3, 0.1, 0.5, 0.6, 0.7, 0.8, 0.9, 1, 2]
}
```

#### Объявим функцию поиска оптимальных гиперпараметров

```python
lr_search = GridSearchCV(
    lr_pipe, 
    lr_param_grid,
    scoring = 'roc_auc',
    cv = 10,
    verbose = 1
)
```
