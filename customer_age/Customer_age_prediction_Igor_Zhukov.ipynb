# Описание задачи

- Сетевой супермаркет внедряет систему компьютерного зрения для обработки фотографий покупателей. Фотофиксация в прикассовой зоне поможет определять возраст клиентов, чтобы:
    - анализировать покупки и предлагать товары, которые могут заинтересовать покупателей этой возрастной группы;
    - контролировать добросовестность кассиров при продаже алкоголя.


  Нужно построить модель, которая по фотографии определит приблизительный возраст человека.
  
  Средняя абсолютная ошибка при определении возраста должна составить **не более 8 лет**
    
  В нашем распоряжении набор фотографий людей с указанием возраста.

## Объявим функции загрузки данных, компиляции и обучения модели

### Объявим функцию загрузки обучающих данных

- при обработке не будем применять вертикальное отражение, так как не предполагается, что камера в супермаркете будет установлена в перевернутом виде;
- для увеличения репрезентативности выборки применим горизонтальное отражение и сдвиг

````python
def load_train(path):
    path = path+'final_files/'

    labels = pd.read_csv('/datasets/faces/labels.csv')                   
    train_datagen = ImageDataGenerator(
        validation_split=0.25,
        rescale=1./255,
        horizontal_flip=True,
        #vertical_flip = True,
        #rotation_range = 90,
        width_shift_range=0.2,
        height_shift_range=0.2
    )


    train_datagen_flow = train_datagen.flow_from_dataframe(
        dataframe = labels,
        directory=path,
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=32,
        class_mode='raw',
        subset='training',
        seed=12345)


    print('path', path)
    
    print('Размер обучающего датагена', 
          len(train_datagen_flow))
    
    return train_datagen_flow
````

### Объявим функцию загрузки тестовых данных 
(без обработки, кроме приведения к единому размеру)

````python
def load_test(path):
    
    path = path+'final_files/'
    labels = pd.read_csv('/datasets/faces/labels.csv')
    validation_datagen = ImageDataGenerator(
        validation_split=0.25,
        rescale=1./255)
    
    #Тут можно применить фильтр для тестирования по возрастам
    #labels = labels[labels['real_age'] >= 72]

    val_datagen_flow = validation_datagen.flow_from_dataframe(
        dataframe = labels,
        directory = path,
        x_col = 'file_name',
        y_col = 'real_age',
        target_size=(224, 224),
        batch_size=32,
        class_mode='raw',
        subset='validation',
        seed=12345)

    print('path', path)
    print('Размер валид. датагена', 
         len(val_datagen_flow))

    
    return val_datagen_flow
````

### Объявим функции конфигурации нейросети ResNet50
- удалим и создадим заново два выходных слоя нейросети: 
   - в качестве предпоследнего слоя установим `GlobalAveragePooling2D` для уменьшения количества признаков и устранения последствий затухания сигнала перед передачей в последний слой;
   - поскольку перед нами стоит задача регрессии (определение возраста) на выходе вместо 1000 будет один нейрон с функцией активации `ReLu`;
    - в качестве функции измерения потерь при оптимизации возьмем **среднеквадратичное отклонение**, а в качестве метрики качества модели - **среднее абсолютное отклонение**;
    - оптимизацию будем проводить с использованием алгоритма `ADAM` с шагом обучения `learning_rate = 0.00003`.

````python
def create_model(input_shape):
    backbone = ResNet50(
        input_shape=input_shape,
        weights='/datasets/keras_models/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5',
        include_top=False)
        
    #backbone.trainable = False
    
    model = Sequential()
    
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(units=1, activation='relu'))
    
    optimizer = Adam(
    learning_rate = 0.00003)    


    model.compile(optimizer=optimizer,
                  loss='mse',
                  metrics=['mae'])
    model.summary()

    return model
````

###  Объявим функцию обучения модели на 61 шаге по умолчанию

````python
def train_model(model,
                train_data, 
                test_data,
                batch_size=None, 
                epochs=61,
                steps_per_epoch=None,
                validation_steps=None
               ):
    
    model.fit(train_data, 
              validation_data=test_data,
              batch_size=batch_size, 
              epochs=epochs,
              steps_per_epoch=steps_per_epoch,
              validation_steps=validation_steps,
              verbose=2,
              shuffle=True)

    return model
````

## Заключение

* Для осуществления задачи по анализу фотографий посетителей супермаркета с целью определения возраста клиентов выбрана модель 50 уровневой остаточной сверточной нейронной сети **ResNet50** из библиотеки `tensorflow.keras`, в которую были **внесены следующие изменения**:
    - **удалены** два выходных слоя библиотечной модели и созданы два новых:
        - **последний** слой из 1000 нейронов **заменен** одним нейроном с нелинейной функцией активации **Rectified Linear Unit** (ReLu), которая преобразует входное значение в значение от 0 до положительной бесконечности, что позволило получить на выходе прогноз возраста человека на фотографии;
        - для предотвращения затухания сигнала и уменьшения количества параметров на входе последнего нейрона, в **предпоследнем** слое применен метод объединения всех массивов признаков сети в единый массив с усреднением значений признаков (**Global Average Pooling**), что позволило сократить количество параметров на входе выходного нейрона до 2049 шт.

* В качестве метода оптимизации обучения использован алгоритм стохастического градиентного спуска, основанный на адаптивной оценке моментов (**ADAM**) с параметром скорости обучения `learning_rate` = **0.00003**.

* В качестве функции измерения потерь при оптимизации использовано **среднеквадратичное отклонение**, а в качестве метрики качества - **среднее абсолютное отклонение**
    
    
  Для обучения и тестирования модели использованы предоставленные **7591** фотографий людей с указанием их возраста. Фотографии в большинстве своем цветные, различного размера и качества, но предварительно обработанные: лицо крупным планом, по центру, повернуто в вертикальное положение, проведена нормализация яркости. В части фотографий обработка проведена некорректно. Распределение возраста в выборке характерно для человеческой популяции (распределение Пуассона), минимальный возраст 1 год, максимальный - 100 лет.
